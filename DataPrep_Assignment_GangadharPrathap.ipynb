# Task 1 - Data Cleaning


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
```

```python
df = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 2, 5],
    'age': [25, np.nan, 35, np.nan, np.nan, 40],
    'city': ['Mumbai', 'Delhi', np.nan, 'Mumbai', 'Delhi', np.nan],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'annual_income': [40000, 80000, 60000, 120000, 80000, 95000]
})
```

```python
df['age'].fillna(df['age'].median(), inplace=True)
df['city'].fillna(df['city'].mode()[0], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("Cleaned Data:")
print(df)
```

# Task 2 — Encoding

```python
# One-Hot Encoding for 'city'
df = pd.get_dummies(df, columns=['city'])
```

```python
# Label Encoding for 'gender'
le = LabelEncoder()
df['gender'] = le.fit_transform(df['gender'])

print("\nEncoded Data:")
print(df)
```

# Task 3 - Scaling

```python
# Min-Max Scaling
minmax_scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[['age', 'annual_income']] = minmax_scaler.fit_transform(df_minmax[['age', 'annual_income']])

print("\nMin-Max Scaled Data:")
print(df_minmax)
```

```python
# Standardization
standard_scaler = StandardScaler()
df_standard = df.copy()
df_standard[['age', 'annual_income']] = standard_scaler.fit_transform(df_standard[['age', 'annual_income']])

print("\nStandardized Data:")
print(df_standard)
```

Explanation:



Use Min-Max Scaling when features need to be bounded (e.g., neural networks).


Use Standardization when data follows a normal distribution or for algorithms like SVM, Logistic Regression.
